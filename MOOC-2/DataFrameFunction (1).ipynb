# PySpark DataFrame Operations on Football Dataset
Using SQL and DataFrame API with World Football dataset from GitHub.

```python
from pyspark.sql import SparkSession
import pyspark.sql.functions as F

spark = SparkSession.builder.appName("Football DataFrame Operations").getOrCreate()
```

```python
url = 'https://raw.githubusercontent.com/footballcsv/world/main/WorldCup.csv'
football_df = spark.read.csv(url, header=True, inferSchema=True)
football_df.show(10)
```

```python
football_df.createOrReplaceTempView('football_table')
```

```python
high_goals_df = spark.sql('SELECT * FROM football_table WHERE Goals > 5')
high_goals_df.show(10)
```

```python
football_df = football_df.select('Year','Team','Goals','Opponent')
football_df = football_df.withColumnRenamed('Goals','GoalsScored')
football_df.sort(F.desc('GoalsScored')).show(10)
```

```python
football_df.groupBy('Team').agg(F.sum('GoalsScored').alias('TotalGoals'), F.max('GoalsScored').alias('MaxGoals')).show(10)
```
